```python
#importing libraries
import pandas as pd
import seaborn as sns

```

```python
#importing dataset from url and assigning it to variable
Url = "https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv"

df = pd.read_csv(Url)
```

```python
#row column counts
rows = len(df.axes[0])
cols = len(df.axes[1])
print(f"There are {rows} rows and {cols} columns in above dataset")
# datatypes
df.dtypes
#info and describe
df.info()
df.describe()
```

```python
#unique counts in location column
umiquevalcount = df.location.nunique()
print(f" There are {umiquevalcount} unique values in  location column")
#max frequency in cotinent
convalcount= df['continent'].mode()[0]
print(f"{convalcount} continent has maximum frequency")

#max and mean values of total_cases
maxoftc = df.total_cases.max()
meanoftc = df.total_cases.mean()
print(f"max of total_cases in above dataset is {maxoftc} and mean is {meanoftc}")

#25%,50% & 75% quartile value in 'total_deaths'
tfval =df.total_deaths.quantile(q=0.25)
ftval =df.total_deaths.quantile(q=0.50)
sfval =df.total_deaths.quantile(q=0.75)

print(f" 25 % quartile value in total_deaths is {tfval} and 50% quartile value is  {ftval} and 75% quartile value is {sfval}")

#continent which has highest human development index
df[df.human_development_index.max() == df['human_development_index']][['continent','human_development_index']].head(1)
#Continent with minimum GDP per capita
df[df['gdp_per_capita'].min() == df['gdp_per_capita']][['continent','gdp_per_capita']].head(1)

```

```python
#filtered data with specific columns


filtered_data = df[['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]
filtered_data.head(5)

#updating data frame so fitting above data into dataframe
df1= df.copy()
df= filtered_data
df
```

```python
#Data cleaning
# Removing duplicate values from filtered data
df.drop_duplicates()
# Finding missing values in column
df.isnull().sum()
# removing observations  where continent values  has null values
df.dropna(subset=['continent'],axis=0,inplace=True)
#filling other null values with 0
df.fillna(0, inplace=True)
df.describe
```

```python
# Convert date column in datetime format
df['date'] = pd.to_datetime(df['date'])

#Create new column month after extracting month data from date column
df['month'] = pd.DatetimeIndex(df['date']).month
```

```python
# Data agrigation
# Find max value in all columns using groupby function on 'continent' column
df.groupby(['continent']).max().reset_index()
# Store the result in a new dataframe named 'df_groupby'
df_groupby = df.groupby(['continent']).max().reset_index()
df_groupby.head(5)
```

```python
 # Feature Engineering
 # Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths']/df_groupby['total_cases']
df_groupby.head()

```

```python
# Data visualization
# Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot

sns.displot(x='gdp_per_capita',data =df_groupby)
```

```python
#Plot a scatter plot of 'total_cases' & 'gdp_per_capita'
sns.scatterplot(x='total_cases',y='gdp_per_capita',data=df_groupby)
```

```python
# Plot Pairplot on df_groupby dataset
sns.pairplot(df_groupby)
```

```python
#Plot a bar plot of 'continent' column with 'total_cases'

sns.catplot(x='continent',y='total_cases',data=df_groupby, kind = 'bar')
```
